```python
import pandas as pd
```

```python
data = pd.read_csv('Collision.csv')
```

```python
data
```

```python
Features = data[["ADDRTYPE","INCDTTM", "JUNCTIONTYPE", "LIGHTCOND", "WEATHER", "ROADCOND"]]
```

```python
Features = Features.drop("INCDTTM", 1)
```

```python
Features_ex = pd.get_dummies(Features)
Features_ex
```

```python
Features_ex.to_csv("Collisions_features.csv")
```

```python
from sklearn.svm import OneClassSVM
```

```python
SVM = OneClassSVM(nu=0.9)
```

```python
Features_ex.head()
```

```python
SVM.fit(Features_ex.head(50000))
```

```python
d = Features_ex.iloc[50000:60000]
```

```python
d
```

```python
SVM.decision_function(d)
```

```python
res = SVM.predict(d)
res
```

```python
res
from sklearn.neighbors.kde import KernelDensity
```
